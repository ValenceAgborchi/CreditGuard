```python
import pandas as pd
import numpy as np

df = pd.read_csv('Loan_Default.csv')
print("Rows: ", df.shape[0])
print("Columns: ", df.shape[1])
```

```python
columns_to_drop = [
'ID',
'year',
'Interest_rate_spread',
'submission_of_application',
'co-applicant_credit_type'

]

df = df.drop(columns=columns_to_drop)
print("Columns remaining: ", df.shape[1])
```

```python
columns_to_fill = [
    'rate_of_interest',
    'Upfront_charges',
    'dtir1',
    'income',

    'property_value',
    'LTV',
    'term' 
]

for column in columns_to_fill:
    median_value = df[column].median()
    df[column] = df[column].fillna(median_value)
    print(f"{column}: filled with median {round(median_value, 2)}")
    
```

```python
text_columns = [
    'loan_type', 'loan_purpose', 'Credit_Worthiness', 'credit_type', 'Region', 'Gender', 'age'
]

df = pd.get_dummies(df, columns=text_columns, drop_first=True)
print("Columns after encoding:", df.shape[1])
```

```python
#Feature 1: Loan Burden relative to income
df['loan_to_income'] = df['loan_amount'] / df['income']

#Feature 2: How much income is left after monthly payment
df['monthly_income'] = df['income'] / 12
df['repayment_capacity'] = df['monthly_income'] - (df['loan_amount'] / df['term'])

print("loan_to_income sample:", df['loan_to_income'].head(3).values)
print("repayment_capacity sample:", df['repayment_capacity'].head(3).values)

```
